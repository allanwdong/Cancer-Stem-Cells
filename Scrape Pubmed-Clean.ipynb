```python
# Rescraping data into this format, test run 

testlst=[]

idlist = idlist_10.split(', ')


def download_single(idnum):
    handle_single = Entrez.efetch(db='pubmed', 
                                      id = idnum, 
                                      retmode='xml',
                                    )
    read = handle_single.read()
    handle_single.close()
    time.sleep(1)
    testlst.append(read)


for i in idlist:
    download_single(i)
    if idlist.index(i) % 2 == 0:
        print(idlist.index(i))
```

```python
# Medium size test
# Going back to batches because sanity check on single download found 1 article/sec = +20 hrs
medtestlst = []

test = [100 * x for x in range(0,4)]

def download_xml(start_index):
    handle_test210 = Entrez.efetch(db='pubmed', 
                                      id = idlist_210, 
                                      retmode='xml',
                                      retstart=start_index,
                                      retmax=100
                                    )
    y = handle_test210.read()
    handle_test210.close()
    time.sleep(10)
    medtestlst.append(y)
    
for i in test:
    download_xml(i)
    print(i/100)
```

```python
# Sanity check on data 

medxml = medtestlst[0]
medsoup = BeautifulSoup(medxml, 'xml')
print(medsoup.find('DateCreated'))
print('\n')
print(medsoup.find('AbstractText').get_text())
```
